```python
counter = 0

for x in df_logins.loc[:, "Location"]:
    loc_list = x.split(", ")
    
    if len(loc_list) == 3:
        df_location_dict = {
        "city" : loc_list[0],
        "state" : loc_list[1],
        "country" : loc_list[2],
        }
    else:
        continue
    
    g = Nominatim()
    coords = g.geocode(df_location_dict)

    coords_dict = {
        "lat" : coords.latitude,
        "long" : coords.longitude,
    }
    print(counter, coords_dict)
    counter += 1
```
